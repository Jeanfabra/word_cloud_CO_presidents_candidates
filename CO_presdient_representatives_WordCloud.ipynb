# What are Word Clouds??
**A word cloud is a collection or cluster of words depicted in different sizes. The bigger and bolder the word apperars, the more often it's mentioned with a given text and the more import it is. For this task, we will be using the Python Library called 'word_cloud' developed by Andreas Mueller**

```python
# Gustavo Petro
website_url = rq.get('https://www.portafolio.co/elecciones-2022/discurso-gustavo-petro-tras-ganar-coalicion-del-pacto-historico-562883')
# Francia Márquez
website_url_2 = rq.get('https://diariocriterio.com/francia-marquez-discurso-candidata/')
# Federico Gutiérrez
website_url_3 = rq.get('https://www.portafolio.co/elecciones-2022/encuesta-centro-nacional-de-consultoria-intencion-de-voto-para-la-presidencia-en-colombia-563120')
# Sergio Fajardo
website_url_4 = rq.get('https://www.eltiempo.com/elecciones-2022/consultas/sergio-fajardo-y-su-discurso-tras-ganar-consultas-presidenciales-658130')
# Rodolfo Hernández
website_url_5 = rq.get('https://elpais.com/internacional/2022-03-09/el-redentor-de-bucaramanga.html', 'html-parser')
# Íngrid Betancourt
website_url_6 = rq.get('https://colombia.as.com/colombia/2021/06/24/actualidad/1624492424_861101.html')
print('URL Requests sent')
```

# Gustavo Petro Word Cloud

```python
# Import BeautifulSoup for HTML structure information from our request
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url.text, 'lxml') # Extracting the data from website_url
data = [element.text for element in soup.find_all('p')] # Retreiving text from 'p' founds
data = str(data) # Converting data to STR
```

```python
!pip -q install wordcloud 
!pip -q install stop-words
print("Libraries installed")
```

```python
from stop_words import get_stop_words # The words we want to ignore

stop_words_es = get_stop_words('es') # Spanish Stopwords
```

```python
from wordcloud import WordCloud, STOPWORDS
print('WordCloud is installed an imported!')
```

```python
stop_words = ['año', 'mil', 'millones', 'pesos', 'ciento', 'El', 'En', 'NAceptar', 'n','n\'','xa0\'', 'tiempo', 'TIEMPO', 'CLARO'] + stop_words_es # Adding aditional stopwords
# Initiate world cloud object

amlo_wc = WordCloud(
        background_color = 'white',
        max_words = 2000,
        stopwords = stop_words)
# Generate the word Cloud
amlo_wc.generate(data)
print('¡Data is generated!')
```

```python
# Display the word cloud]
plt.imshow(amlo_wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
# re-generate the word cloud
amlo_wc.generate(data)

# Display the cloud
fig = plt.figure()
fig.set_figwidth(14) # Set width
fig.set_figheight(18) # Set height

plt.imshow(amlo_wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
print('Gustavo Petro Word Cloud')
```
